```python
import pytesseract
from pdf2image import convert_from_path
import os
import pandas as pd
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
pdf_dir = "/content/drive/MyDrive/invoice_project/invoices"
ocr_output_csv = "/content/drive/MyDrive/invoice_project/ocr_invoice_texts.csv"
```

```python
ocr_results = []
```

```python
for filename in os.listdir(pdf_dir):
    if filename.endswith(".pdf"):
        pdf_path = os.path.join(pdf_dir, filename)

        try:
            # Convert PDF to images (usually 1 page per invoice)
            images = convert_from_path(pdf_path)
            full_text = ""

            for img in images:
                text = pytesseract.image_to_string(img)
                full_text += text + "\n"

            ocr_results.append({
                "filename": filename,
                "extracted_text": full_text.strip()
            })

        except Exception as e:
            print(f"Error processing {filename}: {e}")
```

```python
df_ocr = pd.DataFrame(ocr_results)
df_ocr.to_csv(ocr_output_csv, index=False)
```
